```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_errors_per_epoch_df = pd.read_csv("train_errors.txt", sep=',')
train_errors_per_epoch_df.plot(x ="Epochs", y ="Error")
```

```python
correct_guesses_df = pd.read_csv("correct_guesses.txt", sep=',')
wrong_guesses_df   = pd.read_csv("wrong_guesses.txt"  , sep=',')

plt.title("blue for correct guesses , orange for wrong guesses")
sns.scatterplot(data=correct_guesses_df, x="X1", y="X2", marker='.')
sns.scatterplot(data=wrong_guesses_df,   x="X1", y="X2", marker='*')
```
